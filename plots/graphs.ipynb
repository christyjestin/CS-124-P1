```python
import matplotlib.pyplot as plt
```

```python
x_vals = [128, 256, 512, 1024, 2048, 4096, 8192, 16384, 32768, 65536, 131072, 262144]
avg_sizes_0 = [1.205680, 1.185011, 1.212662, 1.202702, 1.207653, 1.199081, 1.199311, 1.206522, 1.201661, 1.201136, 1.202527, 1.205643]

def guess_0(x):
    return 1.2
```

```python
plt.scatter(x_vals, avg_sizes_0, color = 'orange', label='data')
plt.plot(x_vals, [guess_0(x) for x in x_vals], label='line of best fit')
plt.xscale('log')
plt. ylim(1, 1.4)
plt.legend()
plt.title('Average Tree Size vs. # of Points (rand weights)')
plt.ylabel("average tree size")
plt.xlabel("number of points (log scale)")
```

```python
avg_times_2 = [7.601829, 10.675033, 15.092278, 21.049362, 29.622177, 41.808372, 58.910259, 83.051468, 117.433556, 166.102737, 234.588867, 331.629486]
```

```python
def guess_2(x):
    return (x / (x**0.5)) * 0.65
```

```python
plt.scatter(x_vals, avg_times_2, color = 'orange', label='actual')
plt.plot(x_vals, [guess_2(x) for x in x_vals], label='our estimate')
plt.legend()
plt.title('Average Tree Size vs. # of Points (dim=2)')
plt.ylabel("average tree size")
plt.xlabel("number of points")
```
